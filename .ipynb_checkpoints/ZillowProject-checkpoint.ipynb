### Data Checking and Cleaning the 2016 Property Dataset

```python
plt.figure(figsize=(12,20))
properties16.drop('parcelid',axis=1).notnull().mean().sort_values(ascending = True).plot(kind = 'barh')
```

```python
properties16.drop('parcelid', axis=1).notnull().mean().sort_values(ascending = False)
```

```python
#There are lots 'NaN' data, let's see what is missing
def missing_data():
    return properties16.columns[properties16.isnull().any()].tolist()
column_missing_data = properties16[missing_data()].isnull().sum()
column_missing_data.sort(axis=0, ascending=False, kind='quicksort', 
                         na_position='last', inplace=True)
print (column_missing_data)
print('Total of missing data of 2016 properties:',properties16.isnull().values.sum())
```

```python
#drop the missing value
column_missing_data = properties16.isnull().sum(axis=0).reset_index()
column_missing_data.columns = ['column_name', 'missing_count']
column_missing_data = column_missing_data.ix[column_missing_data['missing_count']>0]
column_missing_data = column_missing_data.sort_values(by='missing_count')
column_missing_data['missing_rate']=column_missing_data['missing_count']/2985217
cutoff=0.9
```

```python
properties16.drop(column_missing_data[(column_missing_data.missing_rate>=cutoff)].column_name.values.tolist(),
                    axis=1,inplace=True)
```
